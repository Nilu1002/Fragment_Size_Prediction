```python
# Import necessary libraries
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import shap

# Load the dataset (Assuming the file is in CSV format)
# For local environment, load the CSV directly
simulated_data = pd.read_csv('coal_blasting_simulated_dataset.csv')  # Update the path as necessary

# Prepare the input (X) and output (y)
X = simulated_data.drop(columns=['Mean Fragment Size (X50) (m)'])  # Input features
y = simulated_data['Mean Fragment Size (X50) (m)']  # Target variable

# Standardizing the data (important for deep learning)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Implementing K-fold Cross-validation (5-fold)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
fold = 1
val_losses = []
test_losses = []
mae_scores = []
r2_scores = []

for train_index, test_index in kf.split(X_scaled):
    print(f"Training on Fold {fold}...")
    
    # Split the data into train and test for this fold
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Build the deep learning model
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(1)  # Output layer for regression
    ])

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Implement early stopping and learning rate scheduler
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    lr_schedule = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-3 * 10**(-epoch / 20))

    # Train the model
    history = model.fit(X_train, y_train, epochs=100, validation_split=0.2, verbose=1,
                        callbacks=[early_stopping, lr_schedule])

    # Evaluate the model on the test data
    test_loss = model.evaluate(X_test, y_test, verbose=1)
    test_losses.append(test_loss)
    
    # Predict on the test data
    y_pred = model.predict(X_test)

    # Calculate MAE and R²
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mae_scores.append(mae)
    r2_scores.append(r2)

    print(f"Test Loss (MSE): {test_loss}, MAE: {mae}, R²: {r2}")

    # Plotting the comparison between actual and predicted fragment sizes for each fold
    plt.scatter(y_test, y_pred)
    plt.xlabel('Actual Mean Fragment Size')
    plt.ylabel('Predicted Mean Fragment Size')
    plt.title(f'Actual vs Predicted Fragment Size (Fold {fold})')
    plt.show()

    fold += 1

# Calculate the average performance metrics across all folds
avg_test_loss = np.mean(test_losses)
avg_mae = np.mean(mae_scores)
avg_r2 = np.mean(r2_scores)

print(f"Average Test Loss (MSE): {avg_test_loss}")
print(f"Average Mean Absolute Error (MAE): {avg_mae}")
print(f"Average R-squared (R²): {avg_r2}")

# Plotting Loss Curves for the last fold
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss during Training (Last Fold)')
plt.ylabel('Loss (MSE)')
plt.xlabel('Epoch')
plt.legend(loc='upper right')
plt.show()

# SHAP (Feature Importance)
explainer = shap.KernelExplainer(model.predict, X_train)
shap_values = explainer.shap_values(X_test)

# Plot SHAP summary for the last fold
shap.summary_plot(shap_values, X_test)

```

```python
import seaborn as sns

# Calculate residuals (errors)
residuals = []
folds = []

for fold, (train_index, test_index) in enumerate(kf.split(X_scaled), 1):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # Train the model (repeat your previous steps here for each fold)
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(1)  # Output layer for regression
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=100, validation_split=0.2, verbose=0)
    
    y_pred = model.predict(X_test)
    
    # Calculate residuals for this fold
    residuals_fold = y_test - y_pred.flatten()
    residuals.extend(residuals_fold)
    folds.extend([fold] * len(residuals_fold))

# Create a DataFrame for residuals and fold numbers
residuals_df = pd.DataFrame({'Residuals': residuals, 'Fold': folds})

# Box plot for residuals across folds
plt.figure(figsize=(8, 6))
sns.boxplot(x='Fold', y='Residuals', data=residuals_df)
plt.title('Residuals Distribution Across Folds')
plt.show()
```

```python
# Collect actual and predicted values across all folds
actuals = []
predictions = []
folds = []

for fold, (train_index, test_index) in enumerate(kf.split(X_scaled), 1):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model = tf.keras.Sequential([
        tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=100, validation_split=0.2, verbose=0)

    y_pred = model.predict(X_test)
    
    # Collect actual and predicted values
    actuals.extend(y_test)
    predictions.extend(y_pred.flatten())
    folds.extend([fold] * len(y_test))

# Create a DataFrame for actual and predicted values
comparison_df = pd.DataFrame({'Actual': actuals, 'Predicted': predictions, 'Fold': folds})

# Box plot comparing actual and predicted values across folds
plt.figure(figsize=(10, 6))
sns.boxplot(x='Fold', y='value', hue='variable',
            data=pd.melt(comparison_df, id_vars='Fold', value_vars=['Actual', 'Predicted']))
plt.title('Actual vs Predicted Values Across Folds')
plt.show()
```

```python
# Import necessary libraries for ML algorithms
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import KFold

# Define the models
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42),
    "SVR": SVR(kernel='rbf', C=1.0, epsilon=0.1)
}

# Implementing K-fold Cross-validation (5-fold) for each ML algorithm
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Store the results for each model
results = {}

for model_name, model in models.items():
    fold = 1
    test_losses = []
    mae_scores = []
    r2_scores = []

    print(f"Training {model_name}...")

    for train_index, test_index in kf.split(X_scaled):
        X_train, X_test = X_scaled[train_index], X_scaled[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Train the model
        model.fit(X_train, y_train)
        
        # Predict on the test data
        y_pred = model.predict(X_test)
        
        # Calculate performance metrics
        test_loss = mean_squared_error(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        test_losses.append(test_loss)
        mae_scores.append(mae)
        r2_scores.append(r2)

        print(f"Fold {fold} - MSE: {test_loss}, MAE: {mae}, R²: {r2}")
        fold += 1

    # Store the average results for each model
    results[model_name] = {
        "Average MSE": np.mean(test_losses),
        "Average MAE": np.mean(mae_scores),
        "Average R²": np.mean(r2_scores)
    }

# Print out the results for each model
for model_name, result in results.items():
    print(f"Results for {model_name}:")
    print(f"  Average MSE: {result['Average MSE']}")
    print(f"  Average MAE: {result['Average MAE']}")
    print(f"  Average R²: {result['Average R²']}")
    print()
```
